（1）公鸡 5 元一只，母鸡 3 元一只，小鸡 1 元三只。
用100元买 100 只鸡，问共有多少种买法？输出所有解。


```python
# 原始代码存在错误，需要修正
solution_count = 0
for cock in range(0,21):  # 公鸡数量从0开始，最多能买20只
    for hen in range(0,34):  # 母鸡数量从0开始，最多能买33只
        chick = 100 - cock - hen  # 小鸡数量
        # 检查小鸡数量是否为正，且是3的倍数（确保小鸡数量为整数）
        if chick >= 0 and chick % 3 == 0 and 5 * cock + 3 * hen + chick // 3 == 100:
            print(f"公鸡: {cock}, 母鸡: {hen}, 小鸡: {chick}")
            solution_count += 1
print(f"共有 {solution_count} 种买法")
```

（2）找出1~n中所有完全数。
【相关定义】完全数：一个数等于其所有真因子之和，例如6=1+2+3。
【输入说明】
输入整数n
【输出说明】
输出完全数及因子
输入示例：
30
输出示例：
6 =1+2+3
28 =1+2+4+7+14

```python
n=int(input("请输入一个整数n: "))

def sum_factors(num):
    factors_sum = 0
    for i in range(1, num):
        if num % i == 0:
            factors_sum += i
    return factors_sum

def print_factors(num):
    factors = []
    for i in range(1, num):
        if num % i == 0:
            factors.append(i)
    return "+".join(map(str, factors))

for i in range(1, n+1):
    if i == sum_factors(i):
        factors_str = print_factors(i)
        print(f"{i} ={factors_str}")
```

（3）输出所有满足以下条件的三位数：
个位、十位、百位互不相同；
数字之和等于数字之积；
该数能被其各位数字之和整除。
并统计这类数的个数。
【输入说明】
无输入
【输出说明】
输出所有结果。

```python
cnt=0
for num in range(100,1000):
    # 获取各位数字
    bai = num // 100
    shi = (num // 10) % 10
    ge = num % 10
    
    # 检查三个数字是否互不相同
    if bai == shi or bai == ge or shi == ge:
        continue
    
    # 计算数字之和与数字之积
    digit_sum = bai + shi + ge
    digit_product = bai * shi * ge
    
    # 检查数字之和是否等于数字之积，且该数能否被各位数字之和整除
    if digit_sum == digit_product and num % digit_sum == 0:
        print(num)
        cnt+=1
print(cnt)
```

（4）找出1~n之间所有幸运数。
【相关定义】
幸运数：各位数字之和能被3整除，且各位数字乘积为偶数。
【输入说明】
输入整数n
【输出说明】
输出所有幸运数及个数。
输入示例：
30
输出示例：
6 12 18 21 24 27 30 
7


```python
n = int(input("请输入一个整数n: "))

def get_digits(num):
    """获取数字的各位数字列表"""
    digits = []
    while num > 0:
        digits.append(num % 10)
        num //= 10
    return digits[::-1]  # 反转得到正常顺序

def is_lucky_number(num):
    """判断一个数是否为幸运数"""
    digits = get_digits(num)
    digit_sum = sum(digits)
    digit_product = 1
    for digit in digits:
        digit_product *= digit
    
    # 检查各位数字之和是否能被3整除，且各位数字乘积是否为偶数
    return digit_sum % 3 == 0 and digit_product % 2 == 0

lucky_numbers = []
for i in range(1, n+1):
    if is_lucky_number(i):
        lucky_numbers.append(i)

# 输出幸运数，以空格分隔
for num in lucky_numbers:
    print(num, end=" ")
print()  # 换行

# 输出幸运数个数
print(len(lucky_numbers))
```

（5）输入一个正整数 n，判断它是否可以表示为两个及以上连续正整数之和。
【输入说明】
输入整数n
【输出说明】
输出所有表示方式
输入示例：
15
输出示例：
1+2+3+4+5=15
4+5+6=15
7+8=15


```python
n = int(input("请输入一个正整数n: "))

# 寻找所有连续正整数序列，使其和等于n
results = []

for start in range(1, n//2 + 1):
    total = 0
    for end in range(start, n//2 + 2):
        total += end
        if total == n and end > start:  # 至少两个数
            sequence = list(range(start, end + 1))
            results.append(sequence)
            break
        elif total > n:
            break

# 输出所有可能的表示方式
for seq in results:
    seq_str = "+".join(map(str, seq))
    print(f"{seq_str}={n}")
```
